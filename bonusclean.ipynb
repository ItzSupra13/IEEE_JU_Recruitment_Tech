#Thought Process [2]:
Cleaning Phone numbers

```python
df1["Leader Phone"] #Zeros at front and + have disappeared it seems
```

```python
def valid(number):
    number = int(number)  # Ensure it's int
    if len(str(number))<=10: #India has 10 digit number, but keep <10 anyways cuz it might help later
        return number
    elif len(str(number))==12: #removing 91 from 91(10 digit number)
        return number%(10**10)
    else:
        return None
df1["Leader Phone"] = df1["Leader Phone"].apply(valid)
```

```python
less_than_ten = df1[df1["Leader Phone"].apply(lambda x: len(str(x)) < 10)]["Leader Phone"]
less_than_ten #20 people have input less than 10 digits in this sheet it seems
```

```python
df2["Candidate's Mobile"] = df2["Candidate's Mobile"].apply(valid)
```

```python
less_than_ten = df2[df2["Candidate's Mobile"].apply(lambda x: len(str(x)) < 10)]["Candidate's Mobile"]
less_than_ten #9 people have input less than 10 digits in this sheet it seems
```

**Limitations**: need to compress df2 onto df1 as df2 is flatenned. This process will delete all team member's names leading to abhorent amount of missing name values, while keeping their mobile no.s and emails intact, as we cant just generate candidates name from member emails can we?  
This goes against our goal of Participant type being both members and leader, and participant name having both leader and member names, but ultimately gets our job donw.  
Maybe portal 1 can improve on its registration form to omit this problem all together (just a suggestion i'm giving)

```python
unique_teams_df1 = set(df1["Team Name"].unique())
print(f"Unique teams in df1: {len(unique_teams_df1)}")

unique_teams_df2 = set(df2["Team Name"].unique())
print(f"Unique teams in df2: {len(unique_teams_df2)}")

common_teams = unique_teams_df1 & unique_teams_df2
print(f"Teams common to both: {len(common_teams)}")

expected_teams = unique_teams_df1.union(unique_teams_df2 - common_teams)
print(f"Expected total unique teams after merge: {len(expected_teams)}")

#Before merging just checking here how many should we expect in the final merged dataset (d1+d2-common)
```

```python
# Step 1: Get team name sets
df1_teams = set(df1["Team Name"].unique())
df2_teams = set(df2["Team Name"].unique())
common_teams = df1_teams & df2_teams
only_df2_teams = df2_teams - df1_teams  # the 83 extra teams

# Step 2: Process teams only in df2 (like we did for common ones)
df2_extra = df2[df2["Team Name"].isin(only_df2_teams)]

extra_rows = []

for team in only_df2_teams:
    team_df = df2_extra[df2_extra["Team Name"] == team]

    leader_row = team_df[team_df["Candidate Type"].str.lower() == "team leader"]
    if leader_row.empty:
        continue  # skip if no leader info

    leader_name = leader_row["Candidate's Name"].values[0]
    leader_email = leader_row["Candidate's Email"].values[0]
    leader_phone = int(leader_row["Candidate's Mobile"].values[0])

    member_rows = team_df[team_df["Candidate's Email"] != leader_email]
    member_emails = member_rows["Candidate's Email"].tolist()

    members_str = ", ".join([leader_email] + member_emails)

    extra_rows.append({
        "Team Name": team,
        "Leader Name": leader_name,
        "Leader Email": leader_email,
        "Leader Phone": leader_phone,
        "Members": members_str
    })

# Step 3: Convert and merge
df2_extra_teams_formatted = pd.DataFrame(extra_rows)

# Step 4: Merge with original df1
final_df = pd.concat([df1, df2_extra_teams_formatted], ignore_index=True)
```
